```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.preprocessing import LabelEncoder


dict_encoders = {}
list_ordered = []
list_not_ordered = []

for col in dataset.columns:
    if dataset[col].dtype.name == "category":
        if dataset[col].cat.ordered:
            list_ordered.append(col)
            dict_encoders[col] = OrdinalEncoder()
            dataset[col] = dict_encoders[col].fit_transform(dataset[[col]])
        else:
            list_not_ordered.append(col)
            dict_encoders[col] = OrdinalEncoder()
            dataset[col] = dict_encoders[col].fit_transform(dataset[[col]])

dataset = dataset.astype("int64")
dataset.head(3)
```

```python
correlations = dataset.corr().unstack().sort_values(ascending=False) 
correlations = pd.DataFrame(correlations).reset_index() 
correlations.columns = ['X', 'y', 'correlation'] 
correlations.query("X == 'mh_diagnos_proffesional' & y != 'mh_diagnos_proffesional'").nlargest(10,columns="correlation")
```

```python
from sklearn.feature_selection import SelectKBest, chi2

y = dataset.mh_diagnos_proffesional
X = dataset.drop(['mh_diagnos_proffesional'],axis=1)

selector = SelectKBest(chi2, k=8)
selector.fit(X,y)
X_new = selector.transform(X)

chi_square_test_result = pd.DataFrame({
    'features': X.columns.values,
     'Scores': selector.scores_,
     'p-values': selector.pvalues_
     }).sort_values(by="Scores").nlargest(8,columns="Scores")
chi_square_test_result
```

```python
selected_features = chi_square_test_result.features.to_list()

selected_features.append("mh_diagnos_proffesional")

selected_features_data = dataset[selected_features]
selected_features_data.head()
```

```python
from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()
X_std = Scaler.fit_transform(selected_features_data)
```

```python
from sklearn.decomposition import PCA


pca = PCA().fit(X_std)
var_exp = pca.explained_variance_ratio_
print(var_exp)

cum_var_exp = np.cumsum(var_exp)
print(cum_var_exp)

eig_vecs = pca.components_
 
Y = PCA(n_components=2).fit(X_std).transform(X_std)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca.fit(X_std) 
pca_scores = pca.transform(X_std) 
```

```python
data_comp = pd.DataFrame(pca.components_,columns=selected_features_data.columns)
plt.figure(figsize=(6,6))
sns.heatmap(data_comp,cmap='flare')
```

```python
%pip install yellowbrick
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

plt.figure(figsize=(8,4))

model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12), timings=False)

visualizer.fit(pca_scores)
visualizer.show()         
```

```python
kmeans_pca = KMeans(n_clusters = 3, init="k-means++")
kmeans_pca.fit(pca_scores)
```

```python
d_segm_pca_kmeans = pd.concat([dataset.reset_index(drop=True),pd.DataFrame(pca_scores)],axis=1)

d_segm_pca_kmeans.columns.values[-3: ] = ['Component 1', 'Component 2', 'Component 3']

d_segm_pca_kmeans['Segment K-means PCA'] = kmeans_pca.labels_
```

```python
d_segm_pca_kmeans.head()
```

```python
d_segm_pca_kmeans['Segment'] = d_segm_pca_kmeans['Segment K-means PCA'].map(
    {0:'First',
    1: 'Second',
    })

d_segm_pca_kmeans.head()
```

```python
from matplotlib.colors import ListedColormap
from numpy import pad

Xax = d_segm_pca_kmeans['Component 1']
Yax = d_segm_pca_kmeans['Component 2']
Zax = d_segm_pca_kmeans['Component 3']

fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(projection='3d')
cmap = ListedColormap(sns.color_palette("Set2"))
ax.scatter(Xax, Yax, Zax, c= d_segm_pca_kmeans["Segment K-means PCA"], cmap=cmap, edgecolor='k', s=100, alpha = 0.7)

ax.grid(False)
ax.legend()
ax.set_title('K-Means Clustering with n=3')
ax.set_xlabel('Component 1', labelpad=20)
ax.set_ylabel('Component 2', labelpad=20)
ax.set_zlabel('Component 3', labelpad=20)
```
